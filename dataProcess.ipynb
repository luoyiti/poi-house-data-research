# 房产数据再整理

```python
import json
```

```python
with open("data/house_poi_data_more_precise.json") as f:
    houses = json.load(f)
```

```python
for house in houses:
    if "-" in house['price_per_meter']:
        value = house['price_per_meter']
        print(value)
        house['price_per_meter'] = str(sum([int(i) for i in value.split("-")]) / 2)
        print(house['price_per_meter'])
```

```python
with open("data/house_poi_data_more_precise.json", "w") as f:
    json.dump(houses, f, ensure_ascii=False, indent = 4)
```

```python
with open("data/house_data.json") as f:
    houses = json.load(f)
for house in houses:
    if "-" in house['price_per_meter']:
        value = house['price_per_meter']
        print(value)
        house['price_per_meter'] = str(sum([int(i) for i in value.split("-")]) / 2)
        print(house['price_per_meter'])
with open("data/house_data.json", "w") as f:
    json.dump(houses, f, ensure_ascii=False, indent = 4)
```
